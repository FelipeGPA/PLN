```
from getpass import getpass

COHERE_API_KEY = getpass()  #Key gerada no site do LLM
```

**Importando as bibliotecas necessárias**

```
from langchain.chains import LLMChain
from langchain.llms import Cohere
from langchain.prompts import PromptTemplate
```

```
template = """Question: {question}

Answer: Let's think step by step."""

prompt = PromptTemplate(template=template, input_variables=["question"])

llm = Cohere(cohere_api_key=COHERE_API_KEY)

llm_chain = LLMChain(prompt=prompt, llm=llm)
```

Teste para validarmos se a comunicação com o LLM foi bem sucedida.

```
question = "What NBA player is the most winner of MVP?"

llm_chain.run(question)
```

Importando as bibliotecas necessárias

```
import logging
from telegram import Update
from telegram.ext import ApplicationBuilder, ContextTypes, CommandHandler
```

```
logging.basicConfig(
    format='%(asctime)s - %(name)s - %(levelname)s - %(message)s',
    level=logging.INFO)
```

```
from typing import Final
from telegram import Update
from telegram.ext import *
import logging
import responses
import nest_asyncio
```

```
TOKEN = getpass()  #Inserindo o TOKEN gerado pelo BOTFather (Telegram)
```

```
logging.basicConfig(format='%(asctime)s - %(name)s - %(levelname)s - %(message)s', level=logging.INFO)
logging.info('Inicializando o bot...')

async def handle_message(update: Update, context: ContextTypes.DEFAULT_TYPE):
  question = str(update.message.text).lower()

  await update.message.reply_text(llm_chain.run(question))

async def error(update, context):
  logging.error(f'Update ({update}) caused error: {context.error}')

def main():
  application = Application.builder().token(TOKEN).build()

  # Mensagens

  application.add_handler(MessageHandler(filters.TEXT & ~filters.COMMAND, handle_message))

  # Rodando o bot
  # O nest_asyncio se faz necessário quando rodamos o bot no Colab

  nest_asyncio.apply()

  application.run_polling(allowed_updates=Update.ALL_TYPES)

if __name__ == "__main__":
    main()
```
